```python
import pandas as pd
import numpy as np
```

```python
confirm_df = pd.read_csv('https://raw.githubusercontent.com/CSSEGISandData/COVID-19/master/csse_covid_19_data/csse_covid_19_time_series/time_series_covid19_confirmed_global.csv')
confirm_df.head()
```

```python
death_df = pd.read_csv('https://raw.githubusercontent.com/CSSEGISandData/COVID-19/master/csse_covid_19_data/csse_covid_19_time_series/time_series_covid19_deaths_global.csv')
death_df.head()
```

```python
death_df.isnull().count()
```

```python
confirm_df.info()
```

```python
confirm_df.count()
```

```python
confirm_df.head()
```

```python
dates = confirm_df.columns[4:]
```

```python
confirm_df_full = confirm_df.melt(
    id_vars=['Province/State', 'Country/Region', 'Lat', 'Long'], 
    value_vars=dates, 
    var_name='Date', 
    value_name='Confirmed'
)
```

```python
confirm_df_full
```

```python
dates = death_df.columns[4:]
```

```python
death_df_full = death_df.melt(
    id_vars=['Province/State', 'Country/Region', 'Lat', 'Long'], 
    value_vars=dates, 
    var_name='Date', 
    value_name='Deaths'
)
```

```python
death_df_full
```

```python
full_df = confirm_df_full.merge(right=death_df_full, how='left', on=['Province/State','Country/Region', 'Lat','Date', 'Long'])
```

```python
full_df
```

```python
full_df['Date'] = pd.to_datetime(full_df['Date'])
full_df.head()
```
